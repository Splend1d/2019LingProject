```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
import numpy as np
import time
```

```python
num_articles = "700"
url = "https://go-api.twreporter.org/v1/posts?where=%7B%22categories%22:%7B%22in%22:%5B%22573177cb8c0c261000b3f6d2%22%5D%7D%7D&limit=" + num_articles + "&offset=0"
```

```python
res = requests.get(url)
reporter_json = json.loads(res.text)
df = pd.DataFrame(reporter_json["records"]) #
article_links = df["slug"] #
article_links
```

```python
print("There are {} articles on TW Reporter.".format(len(article_links)))
```

```python
def article_crawler(idx, link):
    print("Crawling Article No. {}.".format(idx))
    res = requests.get("https://go-api.twreporter.org/v1/posts/" + link + "?full=true") #
    article_json = json.loads(res.text)
    time.sleep(0.1)
    return article_json["record"]

article_dic = {}
for idx, link in enumerate(article_links):
    article_dic[idx] = article_crawler(idx, link)
```

```python
article_df = pd.DataFrame(article_dic).T
article_df
```

```python
# check what's new in articles_df
[col for col in article_df.columns if not col in df.columns]
```

```python
cols = ["tags", "categories", "writters"]

for col in cols:
    print("Mutating {}.".format(col))
    names = []
    for n in article_df[col]:
        if isinstance(n, float):
            name = []
        else:
            name = [i["name"] for i in n]
        names.append(name)   
    article_df[col+"_name"] = names
    
article_df
```

```python
article_df = article_df.assign(content_str = lambda article_df: article_df["content"].map(lambda x: [n["content"][0] for n in x["api_data"] if isinstance(n["content"][0], str)]))
article_df
```
