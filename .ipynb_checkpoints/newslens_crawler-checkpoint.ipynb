```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import time
import json
```

```python
last_page = 70
domain = "https://www.thenewslens.com/review?page="
```

```python
def page_crawler(page):
    url = domain + str(page)
    res = requests.get(url)
    soup = BeautifulSoup(res.text, "lxml")
    
    print("Done Crawling Page {}.".format(page))
    return soup

page_lst = [page_crawler(page) for page in range(1, last_page+1)]
```

```python
def get_article_links(soup):
    article_links = [n.find_all("a")[1].get("href") for n in soup.find_all("div", class_="info-box")] #
    article_links = [n for n in article_links if "/article/" in n] #
    return article_links

article_links_lst = []
for soup in page_lst:
    article_links_lst = article_links_lst + get_article_links(soup)
```

```python
len(article_links_lst)
```

```python
def article_crawler(idx, link): ##
    
    print(idx)
    
    res = requests.get(link)
    soup = BeautifulSoup(res.text, "lxml")

    title = soup.find("div", class_="article-header-container").find("header").text
    article_id = link.replace("https://www.thenewslens.com/article/", "")
        
    article_info = soup.find("div", class_="article-info").text.replace(" ", "")
    date, category = article_info.split(",")
    collaboration = re.match("[A-Za-z]*", date).group()
    date = date.replace(collaboration, "")
    
    WhyNeedKnow = soup.find("header", class_="WhyNeedKnow").find("p").text
    
    content = [n.text for n in soup.find("article").find_all(["h5", "p"])]
    def content_analyzer(n):
        try:
            author_yes = re.match("文：", n[0])
            if author_yes:
                author = n[0].replace("文：", "")
                content_str = " ".join(n[1:])
            else:
                author = ""
                content_str = " ".join(n)
        except:
            author = ""
            content_str = " ".join(n)

        return [re.sub("責任編輯(.*)", "", content_str), author]
    content_str, author = content_analyzer(content)
    
    try:
        tags = [n.get("title") for n in soup.find("ul", class_="tags").find_all("a")]
    except:
        tags = []
        
    time.sleep(0.1)
    
    return {"link": link, "title": title, "article_id": article_id,
            "date": date, "category": category, "collaboration": collaboration,
            "WhyNeedKnow": WhyNeedKnow, "content": content, "content_str": content_str, "tags": tags, "author": author}
    
dic = {}
for idx, link in enumerate(article_links_lst):
    dic[idx] = article_crawler(idx, link)

dic
```

```python
df = pd.DataFrame(dic).T
df
```

```python
with open("newslens.json", "w") as f: #
    json.dump(dic ,f)
```

```python
with open("newslens.json", "r") as f: #
    data = json.load(f)
    
data
```
